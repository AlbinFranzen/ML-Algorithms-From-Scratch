# Polynomial Regression

Polynomial regression is simply linear regression but the predictors are transformed prior to modelling. The formula for simple polynomial regression becomes:

$$y=\theta_0  + \theta_1 x+\theta_2 x^2+...+\theta_d x^d+\epsilon$$

Here we implement it below:

```python
# to make this notebook's output stable across runs
import numpy as np
np.random.seed(42)
```

```python
X = 2 * np.random.rand(100,1)
y = 4 + 4 * X + 2 * X**2 + np.random.randn(100,1)
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.plot(X, y, "b.")
plt.xlabel("$x$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
#plt.axis([0, 2, 0, 15])
plt.show()
```

```python
new_X = np.hstack((X, X**2))
```

```python
from sklearn.linear_model import LinearRegression

poly_reg = LinearRegression()
poly_reg.fit(new_X,y)
```

```python
test_data_x = np.linspace(0,2,100)
transformed_x = np.vstack((test_data_x, test_data_x**2))
test_data_y = poly_reg.predict(transformed_x.T)
```

```python
plt.plot(test_data_x, test_data_y, color="r")
plt.plot(test_data_x, 4+4*test_data_x+2*test_data_x**2, color="g")
plt.plot(X, y, "b.")
plt.xlabel("$x$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.show()
```

```python
print(poly_reg.intercept_)
print(poly_reg.coef_)
```

Our model above is the red line while the green line is the actual polynomial. We can see our model is very close! Polynomial regression can usually reduce variance while increase bias in the model.
